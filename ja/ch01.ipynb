## 00. パタトクカシーー
2つの文字列「パトカー」と「タクシー」の文字を先頭から交互に連結し、文字列「パタトクカシーー」を得よ。

```python
name1 = "パトカー"
name2 = "タクシー"
name3 = ""
for i in range(len(name1)):
  name3 = name3 + name1[i] + name2[i]
print(name3)
```

## 02. 文字列の逆順
文字列"stressed"の文字を逆に（末尾から先頭に向かって）並べた文字列を得よ。

```python
name5 = "stressed"
name6 = ""
for i in range(len(name5)):
  name6 = name6 + name5[len(name5) - 1 - i]
print(name6)
```

## 03. 円周率

```python
text = "Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics."
clean_text = text.replace(",", "").replace(".", "")
words = clean_text.split()
word_lengths = [len(word) for word in words]
print(word_lengths)
```

## 04. 元素記号
"Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."という文を単語に分解し、1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字、それ以外の単語は先頭の2文字を取り出し、取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ。

```python
text = "Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
clean_text = text.replace(".","").replace(",","")
words = clean_text.split()
counter =[1,5,6,7,8,9,15,16,19]
result = {}
for i, word in enumerate(words,start=1):
  if i in counter:
    key = word[0]
  else:
    key = word[0:2]
  result[key] = i+1
print(result)
```

## 05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ。この関数を用い、"I am an NLPer"という文から文字tri-gram、単語bi-gramを得よ。

```python
def get_ngram(target,n):
  result = []
  for i in range(len(target) - n + 1):
    result.append(target[i:i+n])
  return result

text = "I am an NLPer"

text_tri_gram = get_ngram(text.split(),3)
print(text_tri_gram)

text_bi_gram = get_ngram(text.split(),2)
print(text_bi_gram)
```

## 06. 集合
"paraparaparadise"と"paragraph"に含まれる文字bi-gramの集合を、それぞれ, $X$ と $Y$ として求め、$X$ と $Y$ の和集合（$X \cup Y$）、積集合（$X \cap Y$）、差集合（$X \setminus Y$）を求めよ。さらに、'se'というbi-gramがXおよびYに含まれるかどうかを調べよ。

```python
def get_ngram(target,n):
  result = []
  for i in range(len(target) - n + 1):
    result.append(target[i:i+n])
  return result

X = set(get_ngram("paraparaparadise",2))
Y = set(get_ngram("paragraph",2))
print(f"X:{X}")
print(f"Y:{Y}\n")
print(f"和集合:{X|Y}")
print(f"積集合:{X&Y}")
print(f"差集合:{X-Y}")
print(f"'se' in X: {'se' in X}")
print(f"'se' in Y: {'se' in Y}")
```
