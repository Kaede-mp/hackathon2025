## 10. 行数のカウント
ファイルの行数をカウントせよ。確認にはwcコマンドを用いよ。

```python
with open("popular-names.txt", "r", encoding="utf-8") as f:
    lines = len(f.readlines())
    print(f"ファイルの行数: {lines}")
```

## 11. 先頭からN行を出力
ファイルの先頭N行だけを表示せよ。例えば、N=10として先頭10行を表示せよ。確認にはheadコマンドを用いよ。

```python
# 表示したい行数 N を指定
N = 10

with open("popular-names.txt", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        if i >= N:
            break
        # lineには既に改行コードが含まれているため、print側の改行を抑制(end="")
        print(line, end="")
```

```python
!head -n 10 popular-names.txt
```

## 12. 末尾のN行を出力
ファイルの末尾N行だけを表示せよ。例えば、N=10として末尾10行を表示せよ。確認にはtailコマンドを用いよ。

```python
from collections import deque

N = 10

with open("popular-names.txt", "r", encoding="utf-8") as f:
    # maxlen=N を指定することで、末尾のN行だけが効率的に残る
    last_lines = deque(f, maxlen=N)

    # 取得した行を出力（各行に改行コードが含まれるため end="" を指定）
    for line in last_lines:
        print(line, end="")
```

```python
!tail -n 10 popular-names.txt
```

## 13. タブをスペースに置換
ファイルの先頭10行に対して、タブ1文字につきスペース1文字に置換して出力せよ。確認にはsedコマンド、trコマンド、もしくはexpandコマンドなどを用いよ。

```python
N = 10

with open("popular-names.txt", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        # 10行出力したらループを終了
        if i >= N:
            break
        # タブ(\t)をスペースに置換して出力
        print(line.replace("\t", " "), end="")
```

```python
!head -n 10 popular-names.txt | sed 's/\t/ /g'
```

## 14. 1列目を出力
ファイルの先頭10行に対して、各行の1列目だけを抜き出して表示せよ。確認にはcutコマンドなどを用いよ。

```python
N = 10

with open("popular-names.txt", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        if i >= N:
            break
        # 行末の改行を削除してから、タブで分割
        cols = line.strip().split("\t")
        # 1列目（インデックス0）を出力
        print(cols[0])
```

```python
!head -n 10 popular-names.txt | cut -f 1
```

## 15. ファイルをN分割する
ファイルを行単位でN分割し、別のファイルに格納せよ。例えば、N=10としてファイルを10分割せよ。同様の処理をsplitコマンドで実現せよ。

```python
import math

def split_file_into_n(file_path, N):
    # 1. まず全体の行数をカウントする
    with open(file_path, "r", encoding="utf-8") as f:
        total_lines = sum(1 for _ in f)

    # 2. 1ファイルあたりの行数を計算（切り上げ）
    # 例: 2780行を10分割する場合、278行ずつになる
    lines_per_file = math.ceil(total_lines / N)

    # 3. 分割して保存
    with open(file_path, "r", encoding="utf-8") as f:
        for i in range(N):
            # 出力ファイル名 (split_01.txt, split_02.txt...)
            output_filename = f"split_{i+1:02d}.txt"

            with open(output_filename, "w", encoding="utf-8") as out_f:
                for _ in range(lines_per_file):
                    line = f.readline()
                    if not line:
                        break
                    out_f.write(line)

            print(f"{output_filename} を作成しました。")

# 実行 (N=10)
split_file_into_n("popular-names.txt", 10)
```

## 16. ランダムに各行を並び替える
ファイルを行単位でランダムに並び替えよ（注意: 各行の内容は変更せずに並び替えよ）。同様の処理をshufコマンドで実現せよ。

```python
import random

with open("popular-names.txt", "r", encoding="utf-8") as f:
    # 全行をリストとして読み込む
    lines = f.readlines()

# リストの順番をランダムにシャッフル（中身が直接上書きされます）
random.shuffle(lines)

# 確認のため、シャッフル後の先頭5行を出力
print("".join(lines[:5]))

# （※シャッフルした結果を丸ごと保存したい場合）
# with open("shuffled_names.txt", "w", encoding="utf-8") as f:
#     f.write("".join(lines))
```

## 17. １列目の文字列の異なり
1列目の文字列の異なり（文字列の種類）を求めよ。確認にはcut, sort, uniqコマンドを用いよ。

## 18. 各行の1列目の文字列の出現頻度を求め、出現頻度の高い順に並べる
1列目の文字列の出現頻度を求め、出現頻度と名前を出現頻度の多い順に並べて表示せよ。確認にはcut, uniq, sortコマンドを用いよ。

```python
# ユニークな名前を格納するための集合（set）を準備
unique_names = set()

with open("popular-names.txt", "r", encoding="utf-8") as f:
    for line in f:
        # タブで分割して1列目（名前）を取得
        name = line.split("\t")[0]
        # setに保存（重複している場合は自動的に無視されます）
        unique_names.add(name)

# 確認のため、異なり数（種類数）と、名前をアルファベット順にソートして先頭10個を出力
print(f"ユニークな名前の種類数: {len(unique_names)}\n")
print("名前のリスト（先頭10個）:")
print(sorted(unique_names)[:10])
```

```python
# 1列目を切り出し(cut) -> アルファベット順に並び替え(sort) -> 重複削除(uniq)
!cut -f 1 popular-names.txt | sort | uniq | head -n 10

# （※種類数をカウントしたい場合は最後に | wc -l を繋げます）
# !cut -f 1 popular-names.txt | sort | uniq | wc -l
```

## 19. 3列目の数値の降順に各行を並び替える
3列目の数値の逆順でファイルの各行を整列せよ（注意: 各行の内容は変更せずに並び替えよ）。同様の処理をsortコマンドで実現せよ。

```python
with open("popular-names.txt", "r", encoding="utf-8") as f:
    # 全行を読み込む
    lines = f.readlines()

# 3列目の「人数」を数値(int)として解釈し、降順(reverse=True)でソート
# lambda式を使って、各行(line)からソートキーを取り出すロジックを簡潔に記述します
sorted_lines = sorted(
    lines, key=lambda line: int(line.split("\t")[2]), reverse=True
)

# 確認のため、歴代人数トップ10を出力
print("".join(sorted_lines[:10]))

# （※結果をファイルに保存したい場合）
# with open("sorted_names.txt", "w", encoding="utf-8") as f:
#     f.write("".join(sorted_lines))
```

```python
# -k 3 : 3列目を基準にする
# -n   : 文字列としてではなく、数値として並び替える (numeric sort)
# -r   : 逆順／降順にする (reverse)
!sort -n -r -k 3 popular-names.txt | head -n 10
```
